```python
import pandas as pd
```

```python
review=pd.read_csv("dev_with_labels.tsv",delimiter="\t")
review=review.rename(columns={"Text data" : "data"},inplace=False)
review.head()
```

```python
review.describe()
```

```python
from sklearn.model_selection import train_test_split
x=review.data
y=review.Label
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer(stop_words="english",lowercase=False)

#fitting the vectorizer on the training data
vector.fit(x_train)
x_transformed=vector.transform(x_train)   #transform training data into a document term matrix. row->document column->token(word)
x_transformed.toarray()    #converts x_transformed to array. Each element in an array-->count of token

#for test data
x_test_transformed=vector.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_transformed,y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(model.predict(x_test_transformed), y_test))
```
